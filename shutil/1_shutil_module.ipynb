# 1. What is shutil?

`shutil` stands for **Shell Utilities**.

It is a built-in Python module used for:

- Copying files
- Moving files/folders
- Deleting folder trees
- Archiving (zip/tar)
- Disk usage checking
- High-level file operations

Think:

- `os` = low-level file system control  
- `shutil` = high-level ready-made file tools

```python
import shutil
```

# 2. Why shutil?

Without `shutil`, copying files manually is longer.

With `shutil`, one line can do the same job.

```python
# Copy file quickly
shutil.copy("source.txt", "dest.txt")
```

# Recursion

### When a function calls itself repeatedly until it reaches the base case.

### Recursion is a process where a function calls itself repeatedly to solve a smaller version of the problem, and stops when it reaches a base case.


```python
# 4*5  ---> 4+4+4+4+4

def multiply(a,b):

    result = 0
    for i in range(b):
        result += a
    return result
multiply(4,5)  # should return 20
```

```python
# 3 * 4 
# 3 + 3 + 3 + 3
# 3 + 3*3
# 3 + 3 + 2*3
# 3 + 3 + 3 + 1*3

def mul_rec(a,b):
    if b==1:
        return a
    return a + mul_rec(a,b-1)
print(mul_rec(4,5))  # should return 20

# a = 4 b = 3
# is b 1? --> no
# then 4 + mul_rec(4,2)     // its return 4 + 4 + 4
            # a = 4 b = 2
            # is b 1? --> no
            # then 4 + mul_rec(4,1)   // its return 4 + 4
                        # a = 4 b = 1
                        # is b 1? --> yes
                        # then return a



```

```python
# Factorial 

def factorial(n):
    if n== 0:
        return 1
    return n * factorial(n-1)
print(factorial(5))  # should return 120
# 5! = 5 * 4 * 3 * 2 * 1
```

```python
# Palindrome
def palindrome(s):
    if len(s) <= 1:  # Base case: if the string is empty or has one character (odd length or even length)
        print("Palindrome")
    else:
        if s[0] == s[-1]:
            return palindrome(s[1:-1])
        else:
            print("Not a Palindrome")
            
palindrome("madam")
```

```python
# The rabbit problem 
# The problem is to find the number of rabbits after n months, given that each pair of rabbits produces a new pair every month, starting from the second month.
# It is a classic Fibonacci sequence problem, where the number of rabbits in month n is the sum of the number of rabbits in month n-1 and month n-2.

def fibo(n):
    if n==1 or n==0:
        return n
    return fibo(n-1) + fibo(n-2)
print(fibo(12))  # should return 5, as the sequence is 0, 1, 1, 2, 3, 5
```

```python
def memo(m,d):
    if m in d:
        return d[m]
    else:
        d[m]=memo(m-1,d) + memo(m-2,d)
        return d[m]
    
d = {0:0, 1:1}
memo(50,d)


```

```python
# Power set
def power_set(s,index=0,current=""):
    if index == len(s):
        print(f'[{current}]',end='')
        return
    power_set(s,index+1,current) # Exclude the current element
    power_set(s,index+1,current+s[index])

input_str = "123"
power_set(input_str)

```

What is constructor?
Constructor is a method, but it is a special method, it automatically triggered when object is created, it  don't need to call.
* Benefit -  It is used to write configuration related code.

Ex: If god is the programmer, earth is the class, human is the object, then one example of constructor can be death (it will certainly happen) 

* We can't rename constructor name

* Golden rule of object oriented programming - Only object can access the method and attribute of a class.

* Self - Self is the object, using it one method of a class can access other method of same class. It is default parameter. 
* We can use other name instead of self. 








```python
class Fraction:

    # parameterized constructor
    def __init__(self,num,den):
        self.num=num
        self.den=den
    
    def __str__(self):
        return '{}/{}'.format(self.num,self.den)
    
    def __add__(self,other):
        new_num = self.num*other.den + other.num*self.num
        new_den = self.den*other*new_den

        return '{}/{}'.format(new_num,new_den)
    
    def __sub__(self,other):
        new_num = self.num*other.den - other.num*self.num
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)
    
    def __mul__(self,other):
        new_num = self.num*other.num
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)
    
    def __truediv__(self,other):
        new_num = self.num*other.den
        new_den = self.den*other.num

        return '{}/{}'.format(new_num,new_den)
    
    def convertToDecimal(self):
        return self.num/self.den
```

```python
fr1 = Fraction(3,4)
fr2 = Fraction(1,2)
```

```python
fr1.convert_to_decimal()
# 3/4
```

```python
print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)
```
